```python
def outliers(x, sd=3.9):
    low = subset[x].mean() - subset[x].std()*sd
    high = subset[x].mean() + subset[x].std()*sd
    print(low,high)
    mask = np.logical_or(subset[x]> high, subset[x]<low)
    result = subset[x][mask]
    return result

for item in feature_names:
    ans = outliers(item)
    print(ans)
```

```python
def boxoutliers(x):
    q3, q1 = np.percentile(subset[x], [75 ,25])
    IQR = q3 - q1
    lower_fence = q1 - (1.5*IQR)
    upper_fence = q3 + (1.5*IQR)
    print(IQR,lower_fence,upper_fence)
    mask = np.logical_or(subset[x]> upper_fence, subset[x]<lower_fence)
    result = subset[x][mask]
    return result

for item in feature_names:
    ans = boxoutliers(item)
    print(ans)
```

```python
fig = plt.figure(figsize =(10, 7))
fig = plt.boxplot(subset, labels = feature_names)
fig = plt.title("Box plot of each condition",fontsize=14, fontweight='bold')
fig = plt.ylabel("centimeter", fontsize=12)
```
